```python
import numpy as np

def compute_mean(num_list):
    return np.mean(num_list)

print(compute_mean([2, 4, 6, 8]))
```

The code cell above first imports the numpy library, which holds the mean function. Then, it defines the function named compute_mean(), which has the parameter num_list. This parameter should be entered in the form of a list when executing the function. The function returns the mean of the list values. The print statement ensures that the returned value is printed so it can be displayed and read.

Below is the equation of the mean:

$$
\bar{x}=\frac{x_1+x_2+...+x_n}{n}=\frac{1}{n}\left(\sum\limits_{i=1}^{n} x_i \right)
$$
